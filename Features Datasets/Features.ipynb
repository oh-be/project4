def cboe_w(state_new_cases):
    first_Monday_found = False
    week_case_count = 0
    week_case_counts = []
    for index, value in state_new_cases.items():
        index_date = pd.to_datetime(index, format='%m/%d/%Y', 
                                    exact = False)
        index_day_of_week = index_date.day_name()
        if not first_Monday_found and index_day_of_week != 'Monday':
            continue
        first_Monday_found = True
        week_case_count += value
        if index_day_of_week == 'Sunday':
            week_case_counts.append(week_case_count)
            week_case_count = 0
    return week_case_counts

```python
def binning(target, limit):
    # Bucket APPLICATION_TYPE
    bucket = df[target].value_counts()

    # Determine which values to replace if counts are less than `limit`
    replace_list = list(bucket[bucket < limit].index)

    replace_list

    # Replace in dataframe
    for i in replace_list:
        df[target] = df[target].replace(i, "Other")

    return df
```
