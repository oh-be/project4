```python
import pandas as pd
```

```python
#import dataset
data = pd.read_csv('./MLDatasets/rentML.csv')
```

```python
data = data.drop('Unnamed: 0',axis=1)
```

```python
data
```

```python
from pycaret.regression import *
s = setup(data, 
          target = 'rentZestimate',
          session_id=33,
         #ignore_features=['TownCode'],
         #normalize = True,
         #transformation = True,
         #remove_multicollinearity = True,
         #ignore_low_variance = True,
         #combine_rare_levels = True,
         #transform_target = True)
```

```python
best = compare_models()
```

```python
print(best)
```

```python
evaluate_model(best)
```

```python
plot_model(best, plot = 'residuals')
```

```python
plot_model(best, plot = 'feature')
```

```python
predict_model(best)
```

```python
predictions = predict_model(best, data=data)
predictions.head()
```

```python
save_model(best, 'rent_model')
```
